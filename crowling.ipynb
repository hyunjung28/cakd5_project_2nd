```python
# 라이브러리 임포트
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from selenium import Keys
```

```python
# 인기순으로 크롤링

# 당도,산도,바디감, 탄닌의 경우에는 이미지파일로 되어있어 크롤링이 불가하므로 크롤링한 데이터에 수기로 추가해야함

# 드라이버 경로
path = 'C:/Users/kpo01/tool/chromedriver.exe'
driver = webdriver.Chrome(path)
# 브라우저 창을 띄우지 말고 수행하도록 옵션 지정
# options = webdriver.ChromeOptions()
# options.add_argument('headless') 
# driver = webdriver.Chrome(path, options=options)


wine_df = pd.DataFrame(columns = ['이름','영문이름','종류', '나라', '생산지', '제조사', '알코올도수', '포도품종', '가격','평점','주소']) #각 요소로 칼럼을 만듬
idx = 0
for page in range(1,56):
    url = f'https://www.winenara.com/goods/goods_list.php?page={page}&cateCd=001&sort=sellcnt&pageNum=12'
    driver.get(url)
    driver.maximize_window()
    print('현재위치: ', driver.current_url) # 진행 현황 확인
    for i in range(1,13):
        try:
            driver.find_element(By.XPATH, f'//*[@id="contents"]/div/div[2]/div[3]/div[4]/div/div/ul/li[{i}]').click()
            sleep(1)

            # 내용수집
            
            이름 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[1]/h3').text
            영문이름 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[1]/div[1]').text
            종류 = driver.find_element(By.XPATH, '//*[@id="wtype"]').text
            나라 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[1]/li/span[2]').text
            생산지 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[1]').text.split('\n')[1]
            제조사 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[2]').text.split('\n')[1]
            알코올도수 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[3]').text.split('\n')[1]
            포도품종 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[5]').text.split('\n')[1]
            가격 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/dl[3]/dd').text
            평점 = driver.find_element(By.XPATH, '//*[@id="contents"]/div/div[1]/center/div/div[2]/div[1]/div/a/div[2]/div[2]/div[1]').text
            주소 = driver.current_url
            # 데이터프레임에 본문과 카테고리를 행으로 삽입
            wine_df.loc[idx] = [이름,영문이름,종류,나라,생산지,제조사,알코올도수,포도품종,가격,평점,주소]
            idx+=1
            # 뒤로 가기
            driver.back()
            sleep(1)
        except:
            sleep(1)
            # 내용수집
            이름 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[1]/h3').text
            영문이름 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[1]/div[1]').text
            종류 = driver.find_element(By.XPATH, '//*[@id="wtype"]').text
            나라 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[1]/li/span[2]').text
            생산지 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[1]').text.split('\n')[1]
            포도품종 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/ul[2]/li[5]').text.split('\n')[1]
            가격 = driver.find_element(By.XPATH, '//*[@id="frmView"]/div/div/div[2]/dl[3]/dd').text
            주소 = driver.current_url
            # 데이터프레임에 본문과 카테고리를 행으로 삽입
            wine_df.loc[idx] = [이름,영문이름,종류,나라,생산지,'','',포도품종,가격,'',주소]
            idx+=1
            # 뒤로 가기
            driver.back()
            sleep(1)
        

# # CSV파일로 저장하기
# wine_df.to_csv('./wine.csv')
# # 엑셀파일로 저장하기
# wine_df.to_excel('./wine.xlsx')
```

```python
# 2개의 데이터를 통합
wine_df = pd.read_csv('./dataset/wine_df.csv')
wine_df_rc = pd.read_csv('./dataset/wine_by_recommend.csv')
```

```python
# # 인덱스를 순위로서 가공하기 위하여 칼럼 이름 재설정
wine_df.columns = ['인기순위','이름','영문이름','종류', '나라', '생산지', '제조사', '포도품종', '알코올도수' ,'가격','평점','산도','당도','바디감','탄닌','주소']
wine_df_rc.columns = ['추천순위','이름','구매가능여부','주소']
```

```python
# 순위로 가공하기 위해 1을 더함
wine_df['인기순위'] = wine_df['인기순위']+1
wine_df_rc['추천순위'] = wine_df_rc['추천순위']+1
```
